Часто строки и списки используют вместе — из строки с текстом можно составить список слов, поработать с каждым из них и что-то узнать. Составим список слов для дальнейшего анализа из фрагмента текста "Капитанской дочки" Пушкина. Чтобы превратить строку в список, воспользуемся методом `.split('разделитель')`, которым мы пользовались при чтении списка с клавиатуры.

```python
text = "... Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."

words = text.split(' ') # В качестве разделителя воспользуемся пробелом
print(words)            # Посмотрим на результат
```

К сожалению, сейчас слова написаны в разных регистрах: есть с маленькой буквы, а есть и с большой. Исправим это, взяв каждую из строк и превратив её в маленький регистр с помощью метода `.lower()`

```python
text = "... Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."

words = text.split(' ')         # В качестве разделителя воспользуемся пробелом

for i in range(len(words)):
    words[i] = words[i].lower() # Заменим очередное слово на слово только из маленьких букв
print(words)                    # Посмотрим на результат
```

Такая задача — пройти по списку и с каждым элементом что-то сделать — встречается очень часто. Для неё придумали отдельную команду — `map()`.
Функция `map()` принимает имя команды без скобок, которую нужно выполнить с каждым элементом, и структуру данных. Результат работы команды `map()` можно превратить в список. Попробуем напечатать длину всех строк в списке с помощью функции `map()`:

```python
map(len, ['Привет', 'ТЕСТ', 'функции', 'map!'])
```

Как видно из примера выше, функция `map()` возвращает результат, с которым мы пока не можем работать напрямую. Превратим результат её работы в список, и выведем его на экран:

```python
print(list(map(len, ['Привет', 'ТЕСТ', 'функции', 'map!'])))
```

Если нам нужно, чтобы `map()` выполнил с каждым элементом не функцию, а метод — например, метод строк `.lower()` — нам нужно написать так:

```python
print(list(map(str.lower, ['Привет', 'ТЕСТ', 'функции', 'map!'])))
```

``str`` — название типа данных, к которому относится метод. Так как метод `.lower()` относится к строкам, мы пишем слово ``str`` перед ним внутри функции `map()`. Чтобы вспомнить, как называется тип данных, можно воспользоваться уже знакомой вам функцией `type()`: 

```python
print(type('Привет'))
```

Напишем тот же код — преобразование всех букв к нижнему регистру — через `map()`:

```python
text = "... Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."
words = text.split(' ')

# Попросим компьютер пройти по всем словам из words, к каждому слову применить
# строковый метод .lower(), результат работы map() сохранить в виде списка
# и этим новым списком обновить список words
words = list(map(str.lower, words)) 

print(words)
```

К сожалению, теперь в словах нам мешают знаки препинания. Давайте уберём их с помощью метода строк `.strip('символы, которые нужно стереть по бокам слов')`

В прошлый раз мы использовали его для удаления пробелов по бокам от номера банковской карты, но этот метод умеет удалять любые символы, о которых мы попросим:

```python
text = "... Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."
words = text.split(' ')

# Обновим каждое слово, сначала сделав все буквы маленькими,
# потом удалив лишние символы по бокам

for i in range(len(words)):
    words[i] = words[i].lower().strip('.,;:!«»?—') 
    
print(words)
```

Осталось очистить список от пустых слов, которые могли получиться после нашего удаления знаков препинания

```python
text = "... Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."
words = text.split(' ')

for i in range(len(words)):
    words[i] = words[i].lower().strip('.,;:!«»?—') 
    if words[i] == '':           # Если получился пустой элемент — 
        words.remove(words[i])   # удалим его
print(words)                     # Посмотрим на результат
```

Можно заметить, что удаление элемента из списка в процессе перебора элементов приводит к ошибкам. Лучше не удалять элементы из текущего списка, а добавлять в новый.

```python
text = "... Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."
words = text.split(' ')

words_clean = [] # Создадим пустой список для хранения результата

for i in range(len(words)):
    words[i] = words[i].lower().strip('.,;:!«»?—') 
    
    if words[i] != '':                    # Если получилась не пустая строка — 
        words_clean.append(words[i])      # добавим слово к итоговому списку слов

print(words_clean)
```

По аналогии с `.lower()` работает метод `.upper()`, который делает все буквы БОЛЬШИМИ

```python
text = "Родители мои благословили меня. Батюшка сказал мне: «Прощай, Пётр. Служи верно, кому присягнёшь; слушайся начальников; за их лаской не гоняйся; на службу не напрашивайся; от службы не отговаривайся; и помни пословицу: береги платье снову, а честь смолоду». Матушка в слезах наказывала мне беречь моё здоровье, а Савельичу смотреть за дитятей. Надели на меня заячий тулуп, а сверху лисью шубу. Я сел в кибитку с Савельичем и отправился в дорогу, обливаясь слезами."
words = text.split(' ')

i = 0

# Организовать перебор списка можно и через while, но это сложней, 
# ведь здесь нужно будет самому следить за переменной i
while i < len(words):                                                   
    words[i] = words[i].upper().strip('.,;:!«»?—') 
    i += 1                 # Перейдём к следующему слову

print(words)
```

Кроме того, метод строк `.strip()` можно попросить удалить лишние символы только справа или слева, вызвав, соответственно, методы `.rstrip()` и `.lstrip()`:

```python
print('!Удалится справа!'.rstrip('!'))
print('!Удалится слева!'.lstrip('!'))
```

## Повторим все использованные методы.
Методы строк создают новые строки, а **исходную — не изменяют**.

* `.lower()` создаёт новую строчку, где все буквы будут маленькими;
* `.upper()` создаёт новую строчку, где все буквы будут большими;
* `.strip()` создаёт новую строчку, где по бокам исходной строки удалены лишние символы. Без указания аргумента удаляет пробелы, но можно указать строку из символов, которые нужно считать лишними: `.strip('.')` будет считать лишним символом только точки;
* `.lstrip()` и `.rstrip()` работают так же, как и `.strip()`, но удаляют символы только слева или только справа.

Методы можно применять по цепочке друг за другом: `'Hello '.strip().upper()` сначала создаст новую строку без лишнего пробела, а потом превратит все буквы в буквы верхнего регистра: получится новая строка `'HELLO'`.

```python
s = ' .Test. '
print(f'Исходная строка: "{s}"')
print(f'Строка, в которой нет боковых пробелов: "{s.strip()}"')
print(f'Строка, в которой нет боковых пробелов справа: "{s.rstrip()}"')
print(f'Строка, в которой нет боковых пробелов слева: "{s.lstrip()}"')
print(f'Строка, в которой нет боковых пробелов и точек: "{s.strip(" .")}"')
print(f'Строка, в которой все буквы большие: "{s.upper()}"')
print(f'Строка, в которой все буквы маленькие: "{s.lower()}"')
print(f'Все буквы маленькие и нет лишних пробелов: "{s.lower().strip()}"')
print(f'Исходная строка не изменилась: "{s}"')
```
