Мы уже написали несколько программ, в основе которых лежала последовательность действий. Наши программы последовательно выполняли команды одну за другой.

Однако часто какие-либо действия нужно выполнять только при наступлении определенного условия.

 - *Если за окном стало темно*, **мы включаем свет**.
 - *Если земля в горшке с цветком сухая*, **мы поливаем цветок**.
 - *Если в холодильнике нет яиц*,  **мы идем в магазин за яйцами**.
 - *Если на улице идет дождь*, **мы берем зонтик**.
    
Во всех приведенных примерах действие совершалось только при выполнении какого-либо условия.

Если представить выполнение команд нашей программой как путь по какому-то маршруту, то раньше она перемещалась строго по прямой дороге, а теперь на ее пути появляется ответвление. В зависимости от того, выполняется ли условие, программа выбирает маршрут, по которому двигаться дальше.


## Блок-схемы
Чтобы лучше понимать последовательность команд и видеть структуру программы, можно рисовать блок-схемы.

*Блок-схема* — распространенный тип схем, описывающих последовательность команд или процессов, в которых отдельные шаги изображаются в виде блоков различной формы. Эти блоки соединяются между собой линиями, указывающими направление выполнения.
